```python
# Langkah 1: Import library yang diperlukan
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# Langkah 2: Unggah dataset
from google.colab import files
uploaded = files.upload()

# Langkah 3: Memuat dataset
# Membaca file CSV ke dalam DataFrame pandas
import io
data = pd.read_csv(io.BytesIO(uploaded['Position_Salaries.csv']))

# Langkah 4: Tampilkan beberapa baris pertama dataset
print("5 baris pertama dari dataset:")
print(data.head())

# Langkah 5: Pisahkan fitur (X) dan target (y)
# Fitur (X) adalah kolom 'Level', dan target (y) adalah kolom 'Salary'
X = data[['Level']]  # Fitur (Level)
y = data['Salary']   # Target (Salary)

# Langkah 6: Membangun model regresi linear sederhana sebagai pembanding
lin_reg = LinearRegression()
lin_reg.fit(X, y)

# Langkah 7: Membangun model Polynomial Regression (derajat 4)
poly_reg = PolynomialFeatures(degree=4)  # Mengubah derajat polinomial menjadi 4
X_poly = poly_reg.fit_transform(X)

# Langkah 8: Membangun regresi linear untuk model polinomial
lin_reg2 = LinearRegression()
lin_reg2.fit(X_poly, y)

# Langkah 9: Visualisasikan hasil regresi linear sederhana
plt.scatter(X, y, color='blue')
plt.plot(X, lin_reg.predict(X), color='red')
plt.title('Regresi Linear Sederhana')
plt.xlabel('Level Jabatan')
plt.ylabel('Gaji')
plt.show()

# Langkah 10: Visualisasikan hasil Polynomial Regression
plt.scatter(X, y, color='blue')
plt.plot(X, lin_reg2.predict(poly_reg.fit_transform(X)), color='red')
plt.title('Polynomial Regression (Derajat 4)')
plt.xlabel('Level Jabatan')
plt.ylabel('Gaji')
plt.show()

# Langkah 11: Membuat prediksi menggunakan model linear dan polinomial
level = np.array([[6.5]])  # Contoh prediksi untuk level 6.5
# Prediksi dengan model Linear Regression
lin_pred = lin_reg.predict(level)
print(f"Prediksi dengan Linear Regression untuk level 6.5: {lin_pred[0]}")

# Prediksi dengan Polynomial Regression
poly_pred = lin_reg2.predict(poly_reg.fit_transform(level))
print(f"Prediksi dengan Polynomial Regression untuk level 6.5: {poly_pred[0]}")
```
